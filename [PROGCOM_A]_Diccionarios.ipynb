#Diccionarios
Estructura de datos que almacena valores en formato llave:valor. El nombre viene de la analogía con un diccionario del mundo real, en donde la información se encuentra en un formato palabra(llave):definición(valor).

Se define de la siguiente forma
```
d=dict()
```
O de forma alternativa
```
d={}
```
Si se desea definir un diccionario con algunas variables
```
d={"valor1":1}
```
Los diccionarios tienen las siguientes características:
1. No poseen indexación. Para poder acceder a los valores, sólo es necesario colocar el valor entre []
2. Los valores almacenados puede ser cualquier tipo de datos.
3. A diferencia de las listas **no es mutable**.

###EJERCICIO
Generar los siguientes diccionarios:
1. Un diccionario que almacena n nombres de usuarios, sus alturas, pesos, IMC y descripción
2. Un diccionario que almacena n códigos, sus respectivas 5 notas y su promedio
3. Un diccionario que almacena n códigos de productos de una tienda, sus respectivos precios y su % de descuento


````python
import random

def generar_diccionario_usuarios(n):
    # Diccionario de usuarios
    usuarios = {}
    for i in range(n):
        nombre = f'Usuario_{i+1}'
        altura = round(random.uniform(1.5, 2.0), 2)  # Altura entre 1.5 y 2.0 metros
        peso = round(random.uniform(50, 100), 1)  # Peso entre 50 y 100 kg
        imc = round(peso / (altura ** 2), 2)  # Índice de Masa Corporal
        descripcion = f'Usuario de altura {altura} m y peso {peso} kg'
        usuarios[nombre] = {
            'altura': altura,
            'peso': peso,
            'IMC': imc,
            'descripcion': descripcion
        }

    return usuarios

# Generamos el diccionario con n = 5
usuarios = generar_diccionario_usuarios(5)

print(usuarios)
````

````python
import random

def generar_diccionario_notas(n):
    # Diccionario de códigos y notas
    codigos_notas = {}
    for i in range(n):
        codigo = f'Codigo_{i+1}'
        notas = [round(random.uniform(0, 10), 2) for _ in range(5)]  # 5 notas entre 0 y 10
        promedio = round(sum(notas) / len(notas), 2)  # Promedio de notas
        codigos_notas[codigo] = {
            'notas': notas,
            'promedio': promedio
        }

    return codigos_notas

# Generamos el diccionario con n = 5
codigos_notas = generar_diccionario_notas(5)

print(codigos_notas)
````

````python
import random

def generar_diccionario_productos(n):
    # Diccionario de productos
    productos = {}
    for i in range(n):
        codigo_producto = f'Producto_{i+1}'
        precio = round(random.uniform(10, 100), 2)  # Precio entre 10 y 100
        descuento = round(random.uniform(5, 30), 2)  # % de descuento entre 5 y 30%
        productos[codigo_producto] = {
            'precio': precio,
            'descuento': descuento
        }

    return productos

# Generamos el diccionario con n = 5
productos = generar_diccionario_productos(5)

print(productos)
````

##ITERANDO DICCIONARIOS
Los valores de los diccionarios no se recuperan a través de indices, sino que a través de utilizando la llave correspondiente. Esto cambia las reglas al momento de iterar sobre un diccionario, ya que no podemos utilizar un “contador” que tome los valores de los índices como lo hacíamos con las listas (a menos que definamos las llaves como enteros). Es necesario ir iterando sobre las llaves o sobre los valores, esto lo podemos hacer utilizando:

* D.keys(): Retorna las llaves del diccionario D.
* D.values(): Retorna los valores del diccionario D.
Con ayuda de estas funciones, podemos, iterar sobre los diccionarios y realizar operaciones como lo haríamos con una lista.



````python
#Para imprimir las llaves del diccionario
for key in heroes:
  print(key)
````

````python
#para imprimir sólo los valores
for valor in heroes:
  print(heroes[valor])
````

````python
#También podría hacerse de la siguiente manera:
for valor in heroes.values():
  print(valor)
````

````python
#Ahora, si se desea imprimir la llave y su valor a la vez
# Define the dictionary "heroes"
heroes = {
    "Batman": "Bruce Wayne",
    "Superman": "Clark Kent",
    "Wonder Woman": "Diana Prince"
}

#Ahora, si se desea imprimir la llave y su valor a la vez
for llave, valor in heroes.items():
  print(llave, valor)
````

##Métodos
Así como las listas y los conjuntos, los diccionarios también cuentan con métodos. Esta vez, la propuesta es tomar cada uno de los métodos, describir qué hace, y dar un ejemplo:

image.png

1. copy()
Crea una copia superficial del diccionario. Esto significa que se crea un nuevo diccionario que contiene las mismas claves y valores que el diccionario original.

2. clear()
Elimina todos los elementos del diccionario, dejándolo vacío.

3. fromkeys()
Crea un nuevo diccionario con claves de un iterable (como una lista o un conjunto) y un valor opcional que se asigna a todas las claves.

4. items()
Devuelve una vista de los pares clave-valor del diccionario como un objeto de vista de diccionario, que puede ser convertido a una lista.

5. get()
Devuelve el valor de una clave específica. Si la clave no existe, devuelve None o un valor predeterminado si se especifica.

6. keys()
Devuelve una vista de las claves del diccionario. Este objeto se puede convertir a una lista si es necesario.

7. pop()
Elimina y devuelve el valor de una clave específica del diccionario. Si la clave no existe, puede lanzar un error a menos que se proporcione un valor predeterminado.

8. values()
Devuelve una vista de los valores del diccionario. Al igual que keys(), este objeto se puede convertir a una lista.

9. update()
Actualiza el diccionario, agregando pares clave-valor de otro diccionario o iterable. Si una clave ya existe, su valor se actualiza.

10. isetdefault()
Devuelve el valor de una clave específica. Si la clave no existe, se agrega al diccionario con un valor predeterminado y se devuelve ese valor.

11. popitem()
Elimina y devuelve el último par clave-valor agregado al diccionario como una tupla. Si el diccionario está vacío, lanzará un error.
