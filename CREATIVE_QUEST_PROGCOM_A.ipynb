```python

!pip install ipywidgets -q

import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import numpy as np


class Protesis:
    def __init__(self, longitud, diametro):
        self.longitud = longitud
        self.diametro = diametro

    def mostrar_detalles(self):
        return (f"Detalles de la prótesis:\n"
                f"Longitud de la pierna: {self.longitud} cm\n"
                f"Diámetro de la parte afectada: {self.diametro} cm")

    def graficar(self):
        fig, ax = plt.subplots(figsize=(4, 8))

        radius = self.diametro / 2
        length = self.longitud



        theta = np.linspace(0, 2 * np.pi, 100)
        x_ellipse = radius * np.cos(theta)
        y_ellipse = 0.2 * radius * np.sin(theta)


        ax.plot(x_ellipse, y_ellipse + length, color='blue')
        ax.fill_between(x_ellipse, y_ellipse + length, length, color='skyblue', alpha=0.7)


        ax.plot(x_ellipse, y_ellipse, color='blue')
        ax.fill_between(x_ellipse, y_ellipse, 0, color='skyblue', alpha=0.7)


        ax.fill_betweenx(np.linspace(0, length, 100), -radius, radius, color='skyblue', alpha=0.7)
        ax.plot([-radius, -radius], [0, length], color='blue')
        ax.plot([radius, radius], [0, length], color='blue')


        ax.set_xlim(-radius * 1.5, radius * 1.5)
        ax.set_ylim(-radius * 0.5, length + radius * 0.5)
        ax.set_aspect('equal')
        ax.set_xlabel("Diámetro (cm)")
        ax.set_ylabel("Longitud (cm)")
        ax.set_title("Representación gráfica de la prótesis (cilindro)")
        ax.grid(True)

        plt.show()


longitud_input = widgets.FloatText(
    description='Longitud (cm):',
    value=0.0,
    step=0.1,
    style={'description_width': 'initial'}
)

diametro_input = widgets.FloatText(
    description='Diámetro (cm):',
    value=0.0,
    step=0.1,
    style={'description_width': 'initial'}
)

boton = widgets.Button(description="Crear Prótesis")
output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output()
        try:
            longitud = longitud_input.value
            diametro = diametro_input.value

            if longitud <= 0 or diametro <= 0:
                print("❌ Las medidas deben ser valores positivos mayores a 0. Inténtelo de nuevo.")
                return

            protesis_personalizada = Protesis(longitud, diametro)
            print("✅ Prótesis creada con éxito:")
            print(protesis_personalizada.mostrar_detalles())
            protesis_personalizada.graficar()

        except Exception as e:
            print("❌ Error: Por favor ingrese un número válido para las medidas.")

boton.on_click(on_button_click)


display(widgets.VBox([
    widgets.HTML("<h2>Sistema de diseño de prótesis personalizada</h2>"),
    longitud_input,
    diametro_input,
    boton,
    output
]))
```
